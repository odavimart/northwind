```python
orders_details_merged = orders.merge(order_details, on='order_id', how='inner')
orders_customers = orders_details_merged.merge(customers, on='customer_id', how='inner')
orders_customers_products = orders_customers.merge(products, on='product_id', how='inner')
full_data = orders_customers_products.merge(employees, on='employee_id', how='inner')
full_data = full_data.merge(categories, on='category_id', how='inner')
full_data = full_data.merge(suppliers, on='supplier_id', how='inner')
display(full_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Exemplo de dados para fins ilustrativos
data = {
    'product_name': ['Produto A', 'Produto B', 'Produto C'],
    'unit_price': [10, 20, 30],
    'quantity': [100, 200, 150],
    'discount': [0.1, 0.2, 0.15]
}

full_data = pd.DataFrame(data)

# Calcular o total de vendas por produto
full_data['total_sales'] = full_data['unit_price'] * full_data['quantity'] * (1 - full_data['discount'])
total_sales_by_product = full_data.groupby('product_name')['total_sales'].sum().reset_index()
total_sales_by_product.columns = ['Product Name', 'Total Sales']

# Visualizar o total de vendas por produto
plt.figure(figsize=(12, 8))
sns.barplot(data=total_sales_by_product, x='Product Name', y='Total Sales')
plt.xticks(rotation=90)
plt.title('Total de Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Total de Vendas')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dados fictícios para fins ilustrativos
data = {
    'order_date': ['2023-01-15', '2023-01-20', '2023-02-10', '2023-02-22', '2023-03-15', '2023-03-25'],
    'unit_price': [10, 20, 30, 40, 50, 60],
    'quantity': [100, 200, 150, 120, 110, 130],
    'discount': [0.1, 0.2, 0.15, 0.1, 0.05, 0.2],
    'last_name': ['Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia'],
    'employee_id': [1, 2, 3, 4, 5, 6]
}

full_data = pd.DataFrame(data)

# Certificar-se de que a coluna 'order_date' está em formato datetime
full_data['order_date'] = pd.to_datetime(full_data['order_date'])

# Calcular o total de vendas por linha
full_data['total_sales'] = full_data['unit_price'] * full_data['quantity'] * (1 - full_data['discount'])

# Adicionar coluna de mês
full_data['month'] = full_data['order_date'].dt.to_period('M').astype(str)

# Calcular vendas por mês
sales_by_month = full_data.groupby('month')['total_sales'].sum().reset_index()

# Calcular o total de vendas por funcionário
sales_by_employee = full_data.groupby(['employee_id', 'last_name'])['total_sales'].sum().reset_index()

# Ordenar os dados por total de vendas
sales_by_employee = sales_by_employee.sort_values(by='total_sales', ascending=False)

```

# Relatório de Análise de Dados da Northwind Traders

## Introdução

Este relatório apresenta uma análise dos dados da Northwind Traders, focando nos seguintes aspectos:
- Total de vendas por produto
- Vendas por mês
- Desempenho dos funcionários

## Total de Vendas por Produto

Os produtos com maior volume de vendas foram:


```python
# Visualizar vendas por mês
plt.figure(figsize=(12, 8))
sns.barplot(data=total_sales_by_product, x='Product Name', y='Total Sales')
plt.xticks(rotation=90)
plt.title('Total de Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Total de Vendas')
plt.show()
```


## Vendas por Mês

As vendas mensais mostram as seguintes tendências:


```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=sales_by_month, x='month', y='total_sales', marker='o')
plt.title('Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45)
plt.show()
```


## Desempenho dos Funcionários

Os funcionários com melhor desempenho em termos de vendas foram:



```python
plt.figure(figsize=(12, 8))
sns.barplot(data=sales_by_employee, x='total_sales', y='last_name', palette='viridis')
plt.title('Desempenho dos Funcionários')
plt.xlabel('Total de Vendas')
plt.ylabel('Funcionário')
plt.show()
```


## Conclusão

- Total de Vendas por Produto: Identificamos os produtos mais vendidos.
- Vendas por Mês: Observamos tendências e sazonalidades nas vendas.
- Desempenho dos Funcionários: Avaliamos o desempenho de cada funcionário.

Essas análises fornecem insights importantes para a Northwind Traders melhorar suas estratégias de vendas e operação.
